```python
import pandas as pd
crime = pd.read_csv("crime.csv")
crime.columns
```

```python
# List of columns you actually want to keep
keep_cols = ['Year', 'County', 'Violent_sum', 'Property_sum', 'Homicide_sum', 'LTtotal_sum']

# Filter the dataframe to only these columns
crime = crime[keep_cols]

crime = crime.rename(columns={
    'Violent_sum': 'Total_Violent',
    'Property_sum': 'Total_Property',
    'LTtotal_sum': 'Larceny_Theft',
    'ForRape_sum': 'Rape'})

crime = crime[crime['Year'] > 2009].reset_index()
crime
```
